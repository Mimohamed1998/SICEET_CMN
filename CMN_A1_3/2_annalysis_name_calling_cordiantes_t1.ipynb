## Measuring the rate of change of cordinates

```python
data   = pd.DataFrame(data=sample_data_t1);
```

```python
percentageChange_t1 = data.pct_change();
```

```python
percentageChange_t1
```

```python
plt.plot(percentageChange_t1.x1)
plt.title("x1")
plt.show()
plt.plot(percentageChange_t1.y1)
plt.title("y1")
plt.show()
plt.plot(percentageChange_t1.z1)
plt.title("z1")
plt.show()
```

```python
plt.plot(sample_data_t1.x1)
plt.plot(percentageChange_t1.x1)
plt.legend(["x1","X1_rate"])
plt.title("x1")
plt.show()


plt.plot(sample_data_t1.y1)
plt.plot(percentageChange_t1.y1)
plt.legend(["y1","y1_rate"])
plt.title("y1")
plt.show()


plt.plot(sample_data_t1.z1)
plt.plot(percentageChange_t1.z1)
plt.legend(["z1","z1_rate"])
plt.title("z1")
plt.show()
```
