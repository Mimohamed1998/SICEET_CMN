## Import Depedencies

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Obtainig data from children CSV file
sample_data = pd.read_csv('children.csv')
```

```python
sample_data
```

```python
plt.plot(sample_data.x1)
plt.title("x1")
plt.show()
plt.plot(sample_data.y1)
plt.title("y1")
plt.show()
plt.plot(sample_data.z1)
plt.title("z1")
plt.show()
# plt.legend(['x1', 'y1', 'z1'])
# plt.xlabel('frame')
# plt.show()
```

## Measuring the rate of change of cordinates

```python
data   = pd.DataFrame(data=sample_data);
```

```python
percentageChange = data.pct_change();
```

```python
percentageChange
```

```python
plt.plot(percentageChange.x1)
plt.title("x1")
plt.show()
plt.plot(percentageChange.y1)
plt.title("y1")
plt.show()
plt.plot(percentageChange.z1)
plt.title("z1")
plt.show()
```
